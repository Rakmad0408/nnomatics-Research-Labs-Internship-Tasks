# 1. User Login Check

```python
# Defining the Authentication System to check User Credentials
def authentication(username, password):
    if username == "admin" and password == "1234":
        print("Login Successful")
        return True
    else:
        print("Invalid Credentials")


def login():
    while True:
        # Defining the User inputs i.e., Username, Password
        username = input("Enter username: ")
        password = input("Enter the password: ")
        if username == '' or password == '':
            print("Please Enter the Credentials")
        else:
            # Calling the authemticatiion System for checking the User Credentials
            if authentication(username, password):
                break
```

```python
login()
```

# 2. Pass / Fail Analyzer

```python
# Defining the Academic Evaluaion Systems
def evaluation_system(marks):
    pass_count = 0
    fail_count = 0
    for i in marks:
        # students who got marks >= 50 are passed and others are failed
        if i>=50:
            pass_count+=1
        else:
            fail_count+=1

    # printing the total no.of Passed Students and Failed Students
    print(f"Total Passed students are: {pass_count}")
    print(f"Total Failed students are: {fail_count}")
```

```python
marks = [45, 78, 90, 33, 60]

# Calling the evaluation_system for total no.of students who passed and failed
evaluation_system(marks)
```

# 3. Simple Data Cleaner

```python
# Defining the Simple Data Cleaner
def data_cleaner(names):
    for i in range(len(names)):
        # Removing whitespaces in beginning and ending then converting all characters to lowercase
        names[i] = names[i].strip().lower()

    print(names)
```

```python
names = [" Alice ", "bob", " CHARLIE "]

# Calling the Simple Data Cleaner
data_cleaner(names)
```

# 4. Message length Analyzer

```python
# Defining the message length Analyzer
def message_length_analyzer(messages):
    for msg in messages:

        # Threshold for flagging
        LIMIT = 10
        length = len(msg)
        
        # Check if length exceeds the limit
        if length > LIMIT:
            status = "[Flagged: Too Long]"
        else:
            status = "[OK]"
            
        # Print the result
        print(f"'{msg}' - Length: {length} {status}")
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]

# Calling the Message Length Analyzer
message_length_analyzer(messages)
```

# 5. Error Message Detector

```python
# Defining the error messages detector
def error_message_detector(logs):
    # initialyzing error messages counter
    errors = 0
    
    for log in logs:
        # checking for error message
        if log.lower() == 'error':
            # Incrementing the errors counter
            errors+=1
    # printing the total error messages
    print(f"Total Error Count are: {errors}")
```

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# calling the error message detector
error_message_detector(logs)
```
