```python
# Input employee Score data
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}
```

```python
# Calculating the highest score among the employees
highest_score = max(employees.values())
# Declaring a string to print all the employees who have highest score
highest_score_employees = ""
# Looping through the each employee and their score
for employee, score in employees.items():
    if score==highest_score:
        # If employee score matches the highest score add their name to the string
        highest_score_employees=highest_score_employees + employee + ", "

# Printing the output of highest score employees for bonus
print(f"Top Performers Eligible for Bonus: {highest_score_employees[:-2]} (Score:{highest_score})")
```

```python
import string
```

```python
# Input search query
query = "Buy mobile phone buy phone online"
```

```python
# Converting the all query characters to lower case
query = query.lower()
# looping through the each character in the query to remove punctuations
for ch in query:
    if ch in string.punctuation:
        # if their is any punctuation marks then they are replace by the empty string
        query = query.replace(ch, "")
```

```python
# Declaring a empty dictionary to store the frequency/count of the each word
freq = {}
# looping through the each word in the query
for word in query.split():
    # Incrementing the word count by 1 each time if it is already present in the freq else assigning to 1
    freq[word] = freq.get(word, 0) + 1
```

```python
# Output dictionary to store words which has more than 1 occurance
keywords = {}
# Looping through the each word and frequency in the freq dictionary
for word, frequency in freq.items():
    if frequency>1:
        # If frequency is greater than 1 then add it to the keywords Dictionary
        keywords[word] = frequency

# Printing the output dictionary which has words more than 1 occurance
print(keywords)
```

```python
# Input sensor readings
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
# Declaring a dictionary to store valid readings and their hour time
valid_readings = {}
# Looping through the each readings
for i in range(len(sensor_readings)):
    # If the sensor readings are divisible by 2 then they are valid readings
    if sensor_readings[i]%2==0:
        # Store the valid readings in the dictionary along with the hour time
        valid_readings[i] = sensor_readings[i]
```

```python
# printing the output in (hour, reading) format
print(f"Valid Sensor Readings (Hour, Value):\n{list(valid_readings.items())}")
```

```python
# Input employees emails
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
```

```python
# Importing the required libraries
import re
import warnings
warnings.filterwarnings('ignore')
```

```python
# Declaring the domain counting dictionary to store the count of each domain
domain_count = {}
# Looping through the each employee eamil
for email in emails:
    # Finding the domain in the email
    domain = re.findall("[@](\S*)", email)[0]
    # Store the domain name with incrementing the frequency of it for each occurance
    domain_count[domain] = domain_count.get(domain, 0) + 1
```

```python
# Total employees emails length
total_mail = len(emails)
# Looping throught the each domian
for key, value in domain_count.items():
    # Calculating the percentage of the domain used by the employees and prniting them 
    print(f"{key}: {int(value/total_mail*100)}%")
```

```python
# Input Sales data
sales = [1200, 1500, 900, 2200, 1400, 3000]
```

```python
# Looping through the each sale
for i in range(1, len(sales)):
    # Calculating the average sale for each passing day
    avg_sales = sum(sales[:i])/len(sales[:i])
    # Calculating the limit or border of the sale to detect the Spikes
    border = avg_sales*1.3
    # Checking for the Spike
    if sales[i]>border:
        # If there is a spike print the spike day and the sale of the day
        print(f"Day {i+1}: {sales[i]}")
```

```python
# Input User Ids
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
```

```python
# Declaring a dictionary to store the occurance of the each unique Id
dict_ = {}
# Looping through the provided user Ids
for id_ in user_ids:
    # Adding the id to the dictionary and Incrementing the user Id by 1 for each occurance
    dict_[id_] = dict_.get(id_, 0)+1

# Looping through the dictionary for ids having occurance more than 1
for id_, times in dict_.items():
    if times>1:
        # If the occurance of the id is more than 1 then print it
        print(f"{id_} --> {times} times")
```
