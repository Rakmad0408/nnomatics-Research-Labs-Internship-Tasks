```python
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}
```

```python
highest_score = max(employees.values())
highest_score_employees = ""
for employee, score in employees.items():
    if score==highest_score:
        highest_score_employees=highest_score_employees + employee + ", "

print(f"Top Performers Eligible for Bonus: {highest_score_employees[:-2]} (Score:{highest_score})")
```

# 2.

```python
import string
```

```python
query = "Buy mobile phone buy phone online"
```

```python
query = query.lower()
for ch in query:
    if ch in string.punctuation:
        query = query.replace(ch, "")
```

```python
freq = {}
for word in query.split():
    freq[word] = freq.get(word, 0) + 1
```

```python
keywords = {}
for word, frequency in freq.items():
    if frequency>1:
        keywords[word] = frequency

keywords
```

# 4.

```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
```

```python
import re
import warnings
warnings.filterwarnings('ignore')
```

```python
domain_count = {}
for email in emails:
    domain = re.findall("[@](\S*)", email)[0]
    domain_count[domain] = domain_count.get(domain, 0) + 1
```

```python
total_mail = len(emails)
for key, value in domain_count.items():
    print(f"{key}: {int(value/total_mail*100)}%")
```

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
```

```python
for i in range(1, len(sales)):
    avg_sales = sum(sales[:i])/len(sales[:i])
    border = avg_sales*1.3
    if sales[i]>border:
        print(f"Day {i+1}: {sales[i]}")
```
